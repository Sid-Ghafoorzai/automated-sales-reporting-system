```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/raw/rossmann_store_sales.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Date"].min(),df["Date"].max()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.replace(r"(?<!^)(?=[A-Z])", "_", regex=True)
    .str.lower()
)
```

```python
# Shape
print("Shape:", df.shape)

# Date range
print("Date range:", df["date"].min(), "to", df["date"].max())

# Number of stores
print("Stores:", df["store"].nunique())

# Duplicates
print("Duplicate rows:", df.duplicated().sum())
print("Duplicate store-date rows:", df.duplicated(subset=["store", "date"]).sum())

# Missing values
print(df.isna().sum())
```

```python
df.to_csv("data/processed/rossmann_sales_cleaned.csv", index=False)
```
