```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
file_path = "rossmann_store_sales.json"

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "rohit265/rossmann-retail-sales-and-store-performance-data",
    file_path
)

print(df.shape)
df.head()
```

```python
output_path = "rossmann_store_sales.csv"
df.to_csv(output_path, index=False)

print(f"Data saved to {output_path}")
```
